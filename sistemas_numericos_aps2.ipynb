```python
arr = np.array([[761, 820, 3]], dtype=np.float32) 
  
print("Array original:") 
print(arr) 
U, s, V = np.linalg.svd(arr, full_matrices=False) 
print("\nFator do array dado por Decomposição de Valor Singular:") 
print("\nU=", U, "\n\ns=", s, "\n\nV=", V) 
```
